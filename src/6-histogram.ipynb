### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt
import pandas as pd

def create_histogram(data, title="Histogram", xlabel="Value", ylabel="Frequency", color="skyblue"):
    if not isinstance(data, list):
        raise ValueError("Input data must be a list.")
    if not data:
        raise ValueError("Input data list cannot be empty.")

    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=range(min(data), max(data) + 2), color=color, edgecolor='black')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(axis='y', alpha=0.75)
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
    create_histogram(data, title="Distribution of Values", xlabel="Data Points", ylabel="Frequency", color="darkgreen")

    data_empty = []
    try:
        create_histogram(data_empty)
    except ValueError as e:
        print(f"Error: {e}")

    data_not_list = 123
    try:
        create_histogram(data_not_list)
    except ValueError as e:
        print(f"Error: {e}")
```
