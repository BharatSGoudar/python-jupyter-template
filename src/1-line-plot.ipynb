### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import pandas as pd

def create_line_plot_squared():
    x = list(range(11))
    y = [n ** 2 for n in x]
    data = pd.DataFrame({'x': x, 'y': y})

    plt.figure(figsize=(8, 6))
    plt.plot(data['x'], data['y'], color='blue', linestyle='-', linewidth=2, marker='o', markersize=6, markerfacecolor='red')
    plt.title('Line Plot of y = x^2')
    plt.xlabel('x (0 to 10)')
    plt.ylabel('y (x squared)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    create_line_plot_squared()
```
