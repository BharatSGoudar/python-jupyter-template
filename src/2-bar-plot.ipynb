### Exercise 2: Bar Plot with Colors

```python

import matplotlib.pyplot as plt
import pandas as pd

def create_bar_plot(categories, values, title="Bar Plot", xlabel="Categories", ylabel="Values"):
    if not isinstance(categories, list) or not isinstance(values, list):
        raise ValueError("Categories and values must be lists.")
    if len(categories) != len(values):
        raise ValueError("Categories and values lists must have the same length.")
    if not categories:
        raise ValueError("Categories list cannot be empty")

    data = pd.DataFrame({'Category': categories, 'Value': values})
    plt.figure(figsize=(8, 6))
    plt.bar(data['Category'], data['Value'], color=['red', 'green', 'blue', 'purple'])
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    create_bar_plot(categories, values, title="Bar Plot of Categories and Values")

    categories_unequal = ['A', 'B', 'C']
    values_unequal = [1, 2, 3, 4]
    try:
        create_bar_plot(categories_unequal, values_unequal)
    except ValueError as e:
        print(f"Error: {e}")

    categories_empty = []
    values_empty = []
    try:
        create_bar_plot(categories_empty, values_empty)
    except ValueError as e:
        print(f"Error: {e}")
```
