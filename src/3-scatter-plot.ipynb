### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt
import pandas as pd

def create_scatter_plot_with_annotations(x, y, title="Scatter Plot with Annotations", xlabel="X", ylabel="Y"):
    if not isinstance(x, list) or not isinstance(y, list):
        raise ValueError("x and y must be lists.")
    if len(x) != len(y):
        raise ValueError("x and y lists must have the same length.")

    data = pd.DataFrame({'x': x, 'y': y})
    plt.figure(figsize=(8, 6))
    plt.scatter(data['x'], data['y'])
    for i, row in data.iterrows():
        plt.annotate(f"({row['x']}, {row['y']})", (row['x'], row['y']), textcoords="offset points", xytext=(5,5), ha='center')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    x_values = [1, 2, 3, 4, 5]
    y_values = [2, 4, 6, 8, 10]
    create_scatter_plot_with_annotations(x_values, y_values)

    x_unequal = [1, 2, 3, 4]
    y_unequal = [2, 4, 6, 8, 10]
    try:
        create_scatter_plot_with_annotations(x_unequal, y_unequal)
    except ValueError as e:
        print(f"Error: {e}")
```
