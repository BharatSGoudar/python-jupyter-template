### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import pandas as pd

def create_subplots():
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))

    x = list(range(11))
    y = [n ** 2 for n in x]
    axs[0, 0].plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o', markersize=6, markerfacecolor='red')
    axs[0, 0].set_title('Line Plot of y = x^2')
    axs[0, 0].set_xlabel('x')
    axs[0, 0].set_ylabel('y')
    axs[0, 0].grid(True)

    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'purple'])
    axs[0, 1].set_title('Bar Plot')
    axs[0, 1].set_xlabel('Categories')
    axs[0, 1].set_ylabel('Values')

    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]
    axs[1, 0].scatter(x_scatter, y_scatter)
    for i, (x_val, y_val) in enumerate(zip(x_scatter, y_scatter)):
        axs[1, 0].annotate(f"({x_val}, {y_val})", (x_val, y_val), textcoords="offset points", xytext=(5, 5), ha='center')
    axs[1, 0].set_title('Scatter Plot with Annotations')
    axs[1, 0].set_xlabel('X')
    axs[1, 0].set_ylabel('Y')

    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    axs[1, 1].set_title('Pie Chart')

    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    create_subplots()
```
