### Exercise 4: Pie Chart with Percentages

```python

import matplotlib.pyplot as plt
import pandas as pd

def create_pie_chart_with_percentages(labels, sizes, title="Pie Chart"):
    if not isinstance(labels, list) or not isinstance(sizes, list):
        raise ValueError("Labels and sizes must be lists.")
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes lists must have the same length.")
    if not labels:
        raise ValueError("Labels list cannot be empty")

    data = pd.DataFrame({'label': labels, 'size': sizes})

    plt.figure(figsize=(8, 6))
    plt.pie(data['size'], labels=data['label'], autopct='%1.1f%%', startangle=90)
    plt.title(title)
    plt.tight_layout()
    plt.show()

if __name__ == '__main__':
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    create_pie_chart_with_percentages(labels, sizes)

    labels_unequal = ['A', 'B', 'C']
    sizes_unequal = [1, 2, 3, 4]
    try:
        create_pie_chart_with_percentages(labels_unequal, sizes_unequal)
    except ValueError as e:
        print(f"Error: {e}")

    labels_empty = []
    sizes_empty = []
    try:
        create_pie_chart_with_percentages(labels_empty, sizes_empty)
    except ValueError as e:
        print(f"Error: {e}")
```
